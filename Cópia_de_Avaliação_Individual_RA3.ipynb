# Questão 2 - Vale até 1 ponto
Usando o Python (matplotlib, sympy e numpy) trace o gráfico e calcule a área sob a função $F(x) = x^5(2 - x)$ para $0 \leq x \leq 2$. Não esqueça de anotar os pontos, hachurar a área calculada e de imprimir o valor da área encontrada.

```python
import numpy as np
import matplotlib.pyplot as plt


f = lambda x: (x**5) * (2 - x)
g = lambda x: 0
x_1 = 0
x_2 = 2

i_vals = np.linspace(0, 2)
x_vals = np.linspace(x_1, x_2, 400)

plt.plot(i_vals, f(i_vals), color='blue', label=r'$f(x) = x^5(2-x)$')

plt.axvline(x_1, linestyle='--', color='red')
plt.axvline(x_2, linestyle='--', color='green')

def riemann(f, a, b, n):
    r_vals = np.linspace(a, b, n + 1)
    heights = f(r_vals)
    h = (b - a) / n
    return np.sum(heights[:-1] * h)


r_sum = riemann(f, x_1, x_2, 10**8)

print(f"area encontrada por retangulos: {r_sum}")
x = sp.symbols('x')
f1 = (x**5) * (2 - x)

integral = sp.integrate(f1, (x, 0, 2))

integral_num = float(integral.evalf())
print(f"Integral (numérica) de 0 a 2: {integral_num}")
print(f"O erro foi de {integral_num-r_sum}")

plt.fill_between(x_vals, f(x_vals), 0, alpha=0.4, color='grey', label=r'$\approx %.3f$' % r_sum)

plt.legend()
plt.grid(True)
plt.show()
```

Em vermelho temos a linha que define o início do intervalo, e em verde o fim dele.

Eu optei por utilizar o método dos retângulos de Riemann para encontrar a área aproximada da curva, com $10^8$ retângulos, temos o resultado de 3.0476190476190386. Para verificar a validade desse resultado, utilizei ferramentas prontas de integração do sympy, a partir delas, vemos que o erro é de 8.881784197001252e-15, um número extremamente baixo, portanto, o resultado do método dos retângulos é válido.

# Questão 5 - Vale até 1 ponto

Use o método dos discos para determinar o volume do sólido obtido pela rotação da região delimitada pelas funções $y = \sqrt{x}$, $y = 3$ e o próprio eixo $y$ em torno do eixo $y$.


Primeiro, para descobrir o tamanho do sólido, precisamos encontrar a delimitação dele. Como ele gira em torno do eixo y, sua extensão será definida quando y = 3, portanto:

$$3 = \sqrt{x}$$
$$9 = x$$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

f = lambda x: np.sqrt(x)
g = lambda x: np.where(x >= 0, 3, np.nan)


x_1 = 0
x_2 = 9

i_vals = np.linspace(0, 10)
x_vals = np.linspace(x_1, x_2, 4000)

plt.plot(i_vals, f(i_vals), color='blue', label=r'$f(x) = \sqrt{x}$')
plt.plot(i_vals, g(i_vals), color='orange', label=r'$g(x) =3$')


plt.axvline(x_1, linestyle='--', color='red')
plt.axvline(x_2, linestyle='--', color='green')

plt.fill_between(x_vals, f(x_vals), g(x_vals), alpha=0.4, color='grey')

plt.legend()
plt.grid(True)
plt.show()
```

Pelo método de discos, para calcular o volume de um sólido que gira no eixo x, temos:
$$V = \int_a^b \pi [f(x)]^2dx$$
Porém, como estamos rotacionando no eixo y, precisaremos da inversa da função f(x).
A inversa da função f(x) é:

$$f(x)^{-1} = x^2$$

Aplicando a fórmula:

$$V = \int_0^3 \pi [x^2]^2dx$$
$$V = \int_0^3 \pi [x^4]dx$$
Resolvendo a integral, tiramos a constante:
$$V =\pi  \int_0^3 [x^4]dx$$

Usando a regra da potência:

$$\frac{x^5}{5}$$

Usando os intervalos:

$$\frac{3^5}{5} - \frac{0^5}{5}$$
$$\frac{243}{5}$$

Voltando no volume:
$$V =\pi  \frac{243}{5}$$

Portanto, o volume desse sólido de revolução é de $\pi  \frac{243}{5}$.
